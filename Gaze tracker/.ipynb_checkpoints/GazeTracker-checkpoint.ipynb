```python
cap=cv2.VideoCapture(0)

while True:
    _, frame=cap.read()
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    rects=detector(gray, 1)
    
    for (i, rect) in enumerate(rects):
        shape=predictor(gray, rect)
        shape=npshape(shape)
#         for (x, y) in shape:
#             cv2.circle(frame, (x,y), 2, (255,0,0), -1)
    
        mask=np.zeros(frame.shape[:2], dtype=np.uint8)
        mask=mask_of_eye(mask, left, shape)
        mask=mask_of_eye(mask, right, shape)
        
        mask=cv2.dilate(mask, kernel, 5)
                
        eyes=cv2.bitwise_and(frame, frame, mask=mask)
        mask=(eyes==[0,0,0]).all(axis=2)
        eyes[mask]=[255,255,255]
        
        gray_eyes=cv2.cvtColor(eyes, cv2.COLOR_BGR2GRAY)
        
        _,thresh=cv2.threshold(gray_eyes, 60, 255, cv2.THRESH_BINARY)
        thresh = cv2.erode(thresh, None, iterations=2)
        thresh = cv2.dilate(thresh, None, iterations=4)
        thresh = cv2.medianBlur(thresh, 3)
        
        mid = (shape[42][0] + shape[39][0]) // 2
        
        thresh = cv2.bitwise_not(thresh)
        cx1, cy1=contouring(thresh[:, 0:mid], mid, frame)
        cx2, cy2=contouring(thresh[:, mid:], mid, frame, True)
        
        if prev1==None and prev2==None:
            if cx1==None or cx2==None:
                prev1=(0,0)
                prev2=(0,0)
            else:
                prev1=(cx1, cy1)
                prev2=(cx2, cy2)
            
        if cx1==None or cy1==None or cx2==None or cy2==None:
            cx1, cy1=prev1
            cx2, cy2=prev2
        
        if cx1>prev1[0] or cx2>prev2[0]:
            pag.moveRel(-(abs(cx1-prev1[0])*(100)), 0)
        
        if cx1<prev1[0] or cx2<prev2[0]:
            pag.moveRel(abs(cx1-prev1[0])*(100), 0)
            
        if cy1>prev1[1] or cy2>prev2[1]:
            pag.moveRel(0, abs(cy1-prev1[1])*(50))
        
        if cy1<prev1[1] or cy2<prev2[1]:
            pag.moveRel(0, -(abs(cy1-prev1[1])*(50)))
            
        prev1=(cx1, cy1)
        prev2=(cx2, cy2)

        frame=cv2.flip(frame, 1)
        cv2.imshow('res', frame)        
        
    if cv2.waitKey(1) & 0xff==ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```
